$\textbf{METODO DE LA TRANSFORMADA INVERSA}$


El método de la transformada inversa utiliza la distribución acumulada F(x) de la distribución que se va a simular. Puesto que F(x) esta definida en el intervalo (0,1) se puede generar un numero uniforme R y tratatr de determinar el valor de la variable aleatoria pra la cual su distribución acumulada es igual a R, es decir, el valor simulado de la variable aleatoria sigue una distribución de probabilidad f(x), se determina al resolver la siguiente ecuación:

$$
F(x) \ ó \ x = F^{-1}(R)
$$

Para este ejemplo utilizaremos una fucnión con distribución exponencial. De la tabla de distribuciones, tenemos que la función de densidad es:
$$
exp(\lambda) = \lambda e^{-\lambda x}
$$
Y su función de distribución acumulada esta dada por
$$
F(x) = \int_{0}^{x} \lambda e^{-\lambda s} = -\frac{1}{\lambda} \lambda e^{-\lambda s} |_{0}^{x} = -e^{-\lambda x} +1 = 1-e^{-\lambda x}
$$
Ahora, aplicamos el método de la transformada inversa, esto se sigue de la siguiente manera:
$$
1-e^{-\lambda x} = u → e^{-\lambda x} = 1-u = → ln(e^{-\lambda x}) = ln(1-u) → x = -\lambda x =  ln(1-u) → x = - \frac{ ln(1-u)}{\lambda}
$$


$\textbf{PSEUDOCOGIDO}$
1. Generar $u \sim (0,1)$
2. Hacer $x =  \frac{ ln(1-u)}{\lambda}$
3. Devolver x
Para este ejemplo utilizamos un $\lambda = 0.5$

```python
import random as rnd
import numpy as np #Esta libreria incluye algunas funciones matematicas. No necesitamos math
import matplotlib.pyplot as plt
```

```python
def expo(lam,n):
  L = -1/lam
  Lista = []
  for i in range(n):
    #Generamos una uniforme
    u = rnd.random() #Genera números aleatorios entre 0 y 1
    x = L* np.log(u)
    Lista.append(x)#Agregamos los elementos a una lista
  return Lista
```

```python
G = expo(0.5,10000000)
```

```python
mu = np.mean(G)
print(mu)
sigma = np.var(G)
print(sigma)
```

$\textbf{GRAFICA DE LA FUNCIÓN ACUMULATIVA}$

```python
def cdf(muestra):
   n = len(muestra)
   x = np.sort(muestra)
   y = np.arange(0,n)/n
   return x,y
```

```python
G = expo(0.5,1000000)
x,y = cdf(G)
```

```python
plt.step(x,y)
plt.axhline(0.5, color="red")
plt.axvline(np.mean(G),color = "green")
```
