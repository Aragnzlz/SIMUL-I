```python
#Invertir una dunción escalonada
import random as rnd
import numpy as np #Esta libreria incluye algunas funciones matematicas. No necesitamos math
import matplotlib.pyplot as plt
```

```python
def E5(n):
  Lista = []
  for i in range(n):
    #Generamos una uniforme
    u = rnd.random() #Genera números aleatorios entre 0 y 1
    if 0 <= u < 0.15:
      x = 1
    elif 0.15 <= u < 0.20:
      x = 2
    elif 0.20 <= u < 0.55:
      x = 3
    elif 0.55 <= u <= 1:
      x = 4
    Lista.append(x)#Agregamos los elementos a una lista
  return Lista
```

```python
G = E5(100)
```

```python
plt.hist(G, color = "lightblue")
```

```python
plt.plot(G)
```

```python
def cdf(muestra):
   n = len(muestra)
   x = np.sort(muestra)
   y = np.arange(0,n)/n
   return x,y
```

```python
G =E5(100000)
x,y = cdf(G)
```

```python
plt.step(x,y)
plt.axhline(0.5, color="red")
plt.axvline(np.mean(G),color = "green")
```
