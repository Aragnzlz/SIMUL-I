# **MÉTODO DE CONGRUENCIA**

```python
#x_0 = input("Ingrese un valor cualesquiera: ")
#n = input("Número de iteraciones: ")
#c = input("Ingrese la constante c: ")
#a = input("Ingrese la constante a: ")
#m = input("Ingrese el modulo: ")
i = 0
x_n = 6
n = 4
a = 5
c = 5
m = 16
xn = []
axn_c = []
xn_16 = []
mod = []
res = []

for i in range(n):
  x_n1  =  x_n
  x_n = (a*x_n + c) % m
  mod.insert(i,x_n)
  xn.insert(i,x_n1)
  axn_c.insert(i, a*x_n1 + c)
  xn_16.insert(i,x_n1/16)

A = [[i, xn[i], axn_c[i], mod[i], xn_16[i]] for i in range(len(xn))]

head = ["ITER", "XN", "AXN+C","MOD16(AXN+C)","XN/16"]
#A.append(head)
A.insert(0,head)
#IMPRIMIR LA MATRIZ
#for fila in A:
#    print(fila)

                  #FORMATO TABLA
from tabulate import tabulate
# Imprimir los datos en formato de tabla
print(tabulate(A, headers="firstrow", tablefmt="fancy_grid"))
```
