# INCISO A






```python
#Inciso (a) xj+1 = (21xj + 15)mod(31) con x0 = 21
i = 0
x_n = 21
n = 50
a = 21
c = 15
m = 31
xn = []
axn_c = []
xn_mod = []
mod = []
res = []
for i in range(n):
  x_n1  =  x_n
  x_n = (a*x_n + c) % m
  mod.insert(i,x_n)
  xn.insert(i,x_n1)
  axn_c.insert(i, a*x_n1 + c)
  xn_mod.insert(i,x_n1/m)
i = 1
n = len(xn)
for i in range(n):
  if xn[0] == xn[i+1]:
    x = i+1
    del xn[x:n]
    break


# El ciclo de vida esta dado por el tamaño de la lista xn

print(" -----> El ciclo de vida es: ", len(xn), "<-----")

A = [[i, xn[i], axn_c[i], mod[i], xn_mod[i]] for i in range(len(xn))]

head = ["ITER", "XN", "AXN+C","MOD31(AXN+C)","XN/31"]
#A.append(head)
A.insert(0,head)
#IMPRIMIR LA MATRIZ
#for fila in A:
#    print(fila)

                  #FORMATO TABLA
from tabulate import tabulate
# Imprimir los datos en formato de tabla
print(tabulate(A, headers="firstrow", tablefmt="fancy_grid"))
```
