```python
import random as rnd
x = rnd.random()
print(x)
if 0 <= x < 0.5:
  print("SOL")
if 0.5 < x <= 1.0:
  print("AGUILA")
```

```python
def volado(n):
  s = 0
  a = 0
  ps = []
  pa = []
  res =[]
  j = 0
  k = 0
  for i in range(n):
    x = rnd.random()
    if 0 <= x < 0.5:
      #print("S")
      s = s+1
      ps.insert(i,round(s/(i+1),3))
      res.insert(i,"S")
    if 0.5 < x <= 1.0:
      #print("A")
      a = a+1
      ps.insert(i,round(s/(i+1),3))
      res.insert(i,"A")
  print(res)
  return ps
n = 1000
prob = volado(n)
ite = list(range(1,n+1))
print(prob)
print(ite)
```

```python
import numpy as np
np.mean(prob)
```

```python
# prompt: Crear una grafica

import matplotlib.pyplot as plt

plt.plot(ite, prob)

plt.xlabel('Volados')

plt.ylabel('Probabilidad')

plt.title('Probabilidad de sacar sol')

plt.show()
```
