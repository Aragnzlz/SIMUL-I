```python
import random as rnd
import numpy as np #Esta libreria incluye algunas funciones matematicas. No necesitamos math
import matplotlib.pyplot as plt
import math
```

```python
def E3(n):
  Lista = []
  for i in range(n):
    #Generamos una uniforme
    u = rnd.random() #Genera n√∫meros aleatorios entre 0 y 1
    if 0<= u < 1/4:
       x = math.sqrt(25/16 - 2*u) + 5/4
    elif 1/4 <= u <= 1:
       x = 4*u
    else:
       x =0
    Lista.append(x)#Agregamos los elementos a una lista
  return Lista
```

```python
G = E3(100)
```

```python
mu = np.mean(G)
print(mu)
sigma = np.var(G)
print(sigma)
```

```python
plt.hist(G,bins = 200, color = "lightblue")
```

```python
plt.plot(G)
```

```python
def cdf(muestra):
   n = len(muestra)
   x = np.sort(muestra)
   y = np.arange(0,n)/n
   return x,y
```

```python
G = E3(100000)
x,y = cdf(G)
```

```python
plt.step(x,y)
plt.axhline(0.5, color="red")
plt.axvline(np.mean(G),color = "green")
```
