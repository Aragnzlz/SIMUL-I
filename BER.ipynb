```python
import random as rnd
import numpy as np #Esta libreria incluye algunas funciones matematicas. No necesitamos math
import matplotlib.pyplot as plt
```

```python
def B(n,p):
  Lista = []
  for i in range(n):
    #Generamos una uniforme
    u = rnd.random() #Genera nÃºmeros aleatorios entre 0 y 1
    if u >= 1-p:
       x=1
    else:
       x=0
    Lista.append(x)#Agregamos los elementos a una lista
  return Lista
```

```python
G = B(10000000,0.75)
```

```python
mu = np.mean(G)
print(mu)
sigma = np.var(G)
print(sigma)
```

```python
plt.hist(G, color = "lightblue")
```

```python
plt.plot(G)
```

```python
def cdf(muestra):
   n = len(muestra)
   x = np.sort(muestra)
   y = np.arange(0,n)/n
   return x,y
```

```python
x,y = cdf(G)
```

```python
plt.step(x,y)
plt.axhline(0.5, color="red")
plt.axvline(np.mean(G),color = "green")
```
